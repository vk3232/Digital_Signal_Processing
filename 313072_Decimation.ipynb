```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy import signal
from scipy.fft import rfft,rfftfreq
```

```python
N = 250
M = 2

Fs = 1000
F1 = 400
F2 = 75

n = np.arange(0,N)
x = (np.sin(2*np.pi*n*F1/Fs)+np.sin(2*np.pi*n*F2/Fs))

plt.figure(1)
plt.plot(n,x)

X = rfft(x)
f = rfftfreq(N)
plt.figure(2)
plt.plot(Fs*f,np.abs(X))
f
```

```python
y = x[::2] # y is decimal version of x do samples i y are half in number comapare to x

Y = rfft(y)
f = rfftfreq(125) # 1/125 =  250/2 because of the decimation

plt.figure(1)
plt.plot(Fs*f,np.abs(Y)) # Fs is now fs after decimation

# Second  freq of 400hz is allowed to 500-400 = 100Hz is drawback of direct decimation 
```

## Implementing FIR filter of 51 tap using Hamming window firwing signal

```python
h = signal.firwin(51, 250/Fs, window ="hamming")
y1 = signal.lfilter(h,1.0,x)

plt.figure(1)
plt.plot(y1)

Y1 = rfft(y1)
f = rfftfreq(N)

plt.figure(2)
plt.plot(f*Fs,np.abs(Y1))
```

```python
y = y1[::M]

plt.figure(1)
plt.plot(y)

Y= rfft(y)
f = rfftfreq(125)

plt.figure(8)
plt.plot(f*Fs/2,np.abs(Y))
```
