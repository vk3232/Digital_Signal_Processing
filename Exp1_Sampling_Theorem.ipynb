**Importing Necessary Libraries**

```python
import matplotlib.pyplot as plt
from scipy.fftpack import fft
import numpy as np
import math as m
```

**Ploting Sinusoidal Signal**

```python
plt.close('all')
Fs= 200
t= np.arange(0,1,1/Fs)
f1=10
x1= np.sin(2*m.pi*t*f1) #Sine wave generation
plot= plt.figure(1)
plt.subplot(2,1,1)
plt.plot(t,x1)
plt.title('Time Domain');
plt.xlabel('Time(sec)');
plt.ylabel('Amplitude');
plt.show()

n= np.size(t)
fr=(Fs/2)*np.linspace(0,1)
X=fft(x1)
X_m=(2/n)*abs(X)
plt.subplot(2,1,2)
plt.plot(X_m)
plt.title('Frequency Domain');
plt.xlabel('Frequency');
plt.ylabel('Amplitude');
plt.show()
```

**Sampling the Signal**

 Sampled with Fs >= 2f (Aliased free condition)

```python
Fs=50;
f1=20;
n=np.arange(0,Fs,1)
X1= np.sin(2*m.pi*(f1/Fs)*n)
plot=plt.figure(2)
plt.subplot(2,1,1)
plt.stem(n,X1,use_line_collection=True)
plt.title('DT Sinusoidal signal');
plt.xlabel('sample index');
plt.ylabel('Amplitude');
plt.show()

n=np.size(n)
X2=fft(X1)
X2_m=(2/n)*abs(X2)
plt.subplot(2,1,2)
plt.stem(X2_m,use_line_collection=True)
plt.ylabel('Amplitude');
plt.show()
```

Sampled with Fs <= 2f (Under Sampling) / (Aliasing) 

```python
Fs=50;
f1=60;
n=np.arange(0,Fs,1)
X1= np.sin(2*m.pi*(f1/Fs)*n)
plot=plt.figure(2)
plt.subplot(2,1,1)
plt.stem(n,X1,use_line_collection=True)
plt.title('DT Sinusoidal signal');
plt.xlabel('sample index');
plt.ylabel('Amplitude');
plt.show()

n=np.size(n)
X2=fft(X1)
X2_m=(2/n)*abs(X2)
plt.subplot(2,1,2)
plt.stem(X2_m,use_line_collection=True)
plt.show()
```
